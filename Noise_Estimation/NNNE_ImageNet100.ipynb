```python
import numpy as np
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
vgg19_bn_top1 = np.load('vgg19_bn_top1.npy')
vgg19_bn_top5 = np.load('vgg19_bn_top5.npy')
```

```python
vgg19_bn_top1 = vgg19_bn_top1 * 0.01
err_vgg19_bn_top1 = 1 - vgg19_bn_top1
vgg19_bn_top1 = vgg19_bn_top1.tolist()
err_vgg19_bn_top1 = err_vgg19_bn_top1.tolist()
err_mean = np.mean(err_vgg19_bn_top1)
err_var = np.var(err_vgg19_bn_top1)
ax = sns.distplot(err_vgg19_bn_top1)
print(err_mean,err_var)
```

```python
mu = err_mean
sigma = err_var ** 0.5
x = np.array(err_vgg19_bn_top1)
np.savetxt("result.txt", x)
num_bins = 30 #直方图柱子的数量
n, bins, patches = plt.hist(x, num_bins, density=1, alpha=0.75)
y = norm.pdf(bins, mu, sigma)

plt.grid(True)
plt.plot(bins, y, 'r--') #绘制y的曲线 
plt.xlabel('values') #绘制x轴 
plt.ylabel('Probability(%)') #绘制y轴 
plt.title('Histogram : $\mu$=' + str(round(mu,2)) + ' $\sigma=$'+str(round(sigma,2)))  #中文标题 u'xxx' 
#plt.subplots_adjust(left=0.15)#左边距 
#plt.savefig('./save/ImageNet_100_vgg19_bn_top1.jpg',dpi = 600)
plt.show()
```
